```python
cat_vars = [
    'sales_channel',
    'trip_type',
    'flight_day',
    'route',
    'booking_origin'
]
for var in cat_vars:
    df[var] = pd.Categorical(df[var])
```

## Probabilidad de acabar la reserva

```python
sns.barplot(
    data = df,
    x = 'sales_channel',
    y = 'booking_complete',
    estimator = 'mean',
    palette = palette,
    edgecolor = 'black',
    linewidth = .8,
    errorbar = None
)
```
